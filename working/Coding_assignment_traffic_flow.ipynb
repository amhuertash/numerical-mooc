## Simulations Results: Part A

```python
import numpy
from matplotlib import pyplot
%matplotlib inline

#set parameters
vmax = 80 #km/h
L = 11 #km
rho_max = 250 #cars/km
nx = 51 #number of spatial discrete points
nt = 100
dt = 0.001 #hours, time-step size
dx = L/(nx -1) #Spatial grid size (km)

x = numpy.linspace(0,L,nx)
rho0 = numpy.ones(nx)*10
rho0[10:20] = 50
v0 = vmax*(1-rho0/rho_max)
```

```python
#Compute v for anytime in hours
def traffic_flow(rho0, dx, dt, t, rho_max, vmax): #time in hours
    rho = rho0.copy()
    #v = v0.copy()
    nt = int(t / dt)
    #print(nt)
    for n in range (1,nt):
        rho[1:] = rho[1:] - (dt/dx)*vmax*( 1 - 2*(rho[1:]/rho_max))*(rho[1:] - rho[:-1])
    v = vmax*(1-rho/rho_max)
    return v;
    #rho_hist.append(rho.copy())
```

```python
#Minimum velocity in t=0
v = traffic_flow(rho0, dx, dt, (0.0/60), rho_max, vmax)
vmin = (numpy.min(v))/ 3.6
print(vmin)
```

```python
# Plot density in t= 0
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('Initial conditions')
pyplot.xlabel('x (km)')
pyplot.ylabel(r'$\rho$ (cars/s)')
pyplot.grid()
pyplot.plot(x, rho0, color='C0', linestyle='--', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 60);
```

```python
# Plot velocity in t= 0
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('Initial conditions')
pyplot.xlabel('x (km)')
pyplot.ylabel(r'$\upsilon$ (m/s)')
pyplot.grid()
pyplot.plot(x, v, color='C0', linestyle='--', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 30);
```

```python
#Average velocity in t=3 min
v = traffic_flow(rho0, dx, dt, (3/60), rho_max, vmax)
vprom = (numpy.mean(v))/ 3.6
print(vprom)
```

```python
v = v/3.6 #km/h to m/s
```

```python
# Plot velocity in t=3 min
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('Initial conditions')
pyplot.xlabel('x (km)')
pyplot.ylabel(r'$\upsilon$ (m/s)')
pyplot.grid()
pyplot.plot(x, v, color='C0', linestyle='--', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 30);
```
